```python
import pandas as pd
```

```python
# Load the dataset
file_path = "/content/data analysis ticket resolution.csv"
df = pd.read_csv(file_path)
```

```python
# Display the first 5 rows to get a sense of the data
print("Data set:")
print(df.head())
```

Data Cleaning


```python
# Check for missing values in each column
print("Missing Values:")
print(df.isnull().sum())
```

```python
# Check for duplicates
print("Duplicate Rows:", df.duplicated().sum())
```

```python
# Make column names consistent
df.columns = df.columns.str.strip().str.replace(" ", "_").str.lower()
```

```python
# Verify the changes
print("Cleaned Column Names:")
print(df.columns)
```

```python
# Save the cleaned data
df.to_csv("Cleaned_Helpdesk_Tickets.csv", index=False)
print("Cleaned data saved as 'Cleaned_Helpdesk_Tickets.csv'")
```

Exploratory Data Analysis

```python
# Ticket distribution by priority
priority_counts = df['priority'].value_counts()
```

```python
print("Ticket Distribution by Priority:")
print(priority_counts)
```

```python
# Visualize the distribution using a bar chart
import matplotlib.pyplot as plt
priority_counts.plot(kind='bar', color='skyblue')
plt.title('Ticket Distribution by Priority')
plt.xlabel('Priority')
plt.ylabel('Number of Tickets')
plt.show()
```

```python
# Create a dictionary for mapping satisfaction levels to numeric values
satisfaction_map = {
    'Unsatisfied': 1,
    'Satisfied': 2,
    'Highly satisfied': 3,
    'Unknown': 0
}
```

```python
# Map the satisfaction column to numeric values
df['satisfaction'] = df['satisfaction'].map(satisfaction_map)
```

```python
# Now calculate the average satisfaction score
avg_satisfaction = df['satisfaction'].mean()
```

```python
# Display the result with two decimal places
print("Average Customer Satisfaction:", round(avg_satisfaction, 3))
```

Analyze the number of tickets categorized by severity.

```python
# Ticket distribution by severity
severity_counts = df['severity'].value_counts()
```

```python
print("Ticket Distribution by Severity:")
print(severity_counts)
```

```python
# Visualize the distribution using a pie chart with improved label readability
plt.figure(figsize=(6, 6))  # Increase the figure size for better clarity
severity_counts.plot(kind='pie',
                     autopct='%1.1f%%',
                     startangle=90,
                     labels=severity_counts.index,
                     pctdistance=0.85,
                     colors=['lightcoral', 'skyblue', 'gold', 'lightgreen'])

# Title for the pie chart
plt.title('Ticket Distribution by Severity')

# Show the chart
plt.show()
```

```python
# Ticket distribution by type
ticket_type_counts = df['tickettype'].value_counts()

print("\n Ticket Distribution by Ticket Type:")
print(ticket_type_counts)

# Visualize the distribution using a horizontal bar chart
ticket_type_counts.plot(kind='barh', color='lightcoral')
plt.title('Ticket Distribution by Type')
plt.xlabel('Number of Tickets')
plt.ylabel('Ticket Type')
plt.show()
```

```python
# Scatter plot for satisfaction vs severity
plt.scatter(df['severity'], df['satisfaction'], alpha=0.5)
plt.title('Satisfaction vs Severity')
plt.xlabel('Severity')
plt.ylabel('Satisfaction')
plt.show()
```
